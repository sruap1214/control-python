## 1. Mapa de polos y zeros

El primer paso antes de mirar el diagrama del lugar geometrico de las raices es mirar en donde se ubican los polos y zeros de una funcion de transferencia. Para lo anterior se parte del siguiente ejemplo:

$$
H(s) = \frac{1}{3s^2 + 4s + 1}
$$

```python
import control as ct

# Define the system
sys = ct.tf([1], [3, 4, 1])

# Compute the poles and zeros
ct.pole_zero_plot(sys)
```

Notese que un sistema de segundo orden, con una realimentacion de esta forma nunca se va a desestabilizar para ningun valor de K. Que pasa para un sistema de tercer orden?

```python
sys1 = ct.tf([1], [1, 1])
sys2 = ct.tf([1], [1, 2])
sys3 = ct.tf([1], [1, 3])

# Series
sys = sys1 * sys2 * sys3
print(sys)

# Root locus
ct.root_locus_plot(sys)
```
